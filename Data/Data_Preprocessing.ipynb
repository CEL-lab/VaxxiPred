```python
import pandas as pd

# Load the Excel file into a DataFrame
file_path = 'Final Candidate Protiens from Research.xlsx'
df_candidates = pd.read_excel(file_path)
```

```python
df_candidates
```

```python
# Count rows with FASTA header (sequences that contain lines starting with '>')
fasta_header_count = df_candidates['Sequences'].apply(lambda x: any(line.startswith('>') for line in str(x).splitlines())).sum()
print(f"Number of sequences with FASTA headers: {fasta_header_count}")
```

```python

# Remove FASTA header lines, keeping only the sequences
df_candidates['Sequences'] = df_candidates['Sequences'].apply(lambda x: '\n'.join(line for line in str(x).splitlines() if not line.startswith('>')))

# Display the cleaned DataFrame
print(df_candidates[['Protien Number', 'Sequences']].head())
```

```python
# Remove empty rows from the DataFrame
df_candidates = df_candidates.dropna(how='all')

# Save the cleaned DataFrame to a CSV file
df_candidates.to_csv('cleaned_candidates.csv', index=False)
```

```python
df_vaccine_candidate = pd.read_csv('cleaned_candidates.csv')
df_vaccine_candidate
```

```python
# Remove rows with NaN values in the 'Protien Number' column
df_vaccine_candidate = df_vaccine_candidate.dropna(subset=['Protien Number']).reset_index(drop=True)

# Remove newline characters within the sequences using .loc to avoid warnings
df_vaccine_candidate.loc[:, 'Sequences'] = df_vaccine_candidate['Sequences'].str.replace('\n', '', regex=True)

# Save the cleaned DataFrame to a new CSV file
df_vaccine_candidate.to_csv('cleaned_vaccine_candidates.csv', index=False)

# Display the cleaned DataFrame
print(df_vaccine_candidate)
```

```python
import pandas as pd
from Bio import SeqIO

# List of FASTA file paths
fasta_files = [
    'uniprotkb_Aeromonas_hydrophila_AND_anno_2024_10_30.fasta',
    'uniprotkb_Edwardsiella_ictaluri_AND_ann_2024_10_30.fasta',
    'uniprotkb_Flavobacterium_columnare_AND_2024_10_30.fasta'
]

# Initialize an empty list to collect data
data = []

# Parse each FASTA file and extract Protein Number and Sequence
for file in fasta_files:
    for record in SeqIO.parse(file, 'fasta'):
        # Extract Protein Number from header
        protein_number = record.id.split('|')[1] if '|' in record.id else record.id
        # Extract sequence
        sequence = str(record.seq)
        # Append to data list
        data.append([protein_number, sequence])

# Create a DataFrame from the data list
df_sequences = pd.DataFrame(data, columns=['Protein Number', 'Sequence'])

# Save DataFrame to a CSV file
df_sequences.to_csv('all_sequences.csv', index=False)

# Display the DataFrame
print(df_sequences)
```
